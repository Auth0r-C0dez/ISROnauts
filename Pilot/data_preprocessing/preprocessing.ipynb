```python
import os
import h5py
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Folder in Drive
folder_path = '/content/drive/My Drive/ISROnauts.AI/ISROnauts.AI/Jun25_132269'

# Bands you want to extract and resize
bands_to_extract = ['IMG_VIS', 'IMG_MIR', 'IMG_SWIR', 'IMG_WV', 'IMG_TIR1', 'IMG_TIR2']

# Where to save output images
output_dir = '/content/resized_all_bands'
os.makedirs(output_dir, exist_ok=True)

# Dictionary to hold resized images: {file_name: {band: image_array}}
all_resized_images = {}

for file_name in os.listdir(folder_path):
    if file_name.endswith('.h5'):
        file_path = os.path.join(folder_path, file_name)
        resized_band_images = {}

        with h5py.File(file_path, 'r') as f:
            for band in bands_to_extract:
                if band in f:
                    image = f[band][0]  # Remove 1st dim (shape becomes 3207√ó3062)

                    # Normalize to 0-255 for visualization
                    norm_img = (255 * (image - np.min(image)) / np.ptp(image)).astype(np.uint8)

                    # Convert to PIL, resize
                    img = Image.fromarray(norm_img)
                    img_resized = img.resize((128, 128))

                    # Save as array and image
                    resized_band_images[band] = np.array(img_resized)

                    # Save to disk as PNG
                    save_name = f"{file_name.replace('.h5', '')}_{band}_128x128.png"
                    img_resized.save(os.path.join(output_dir, save_name))

        all_resized_images[file_name] = resized_band_images
        print(f"‚úî Resized all bands from {file_name}")

# Optional: Display one example
example_file = list(all_resized_images.keys())[0]
example_band = 'IMG_VIS'
plt.imshow(all_resized_images[example_file][example_band], cmap='gray')
plt.title(f"{example_file} - {example_band}")
plt.axis('off')
plt.show()
```

```python
import os
import h5py
import numpy as np
from PIL import Image
from google.colab import drive

# Mount your Google Drive
drive.mount('/content/drive')

# Main output folder in Drive
base_output_dir = '/content/drive/My Drive/ISROnauts.AI/Resized_128x128'
os.makedirs(base_output_dir, exist_ok=True)

# Bands to extract
bands_to_extract = ['IMG_VIS', 'IMG_MIR', 'IMG_SWIR', 'IMG_WV', 'IMG_TIR1', 'IMG_TIR2']

# Your input folder path
folder_path = '/content/drive/My Drive/ISROnauts.AI/ISROnauts.AI/Jun25_132269'

for file_name in os.listdir(folder_path):
    if file_name.endswith('.h5'):
        file_path = os.path.join(folder_path, file_name)

        # Make a subfolder for this file
        file_id = file_name.replace('.h5', '')
        subfolder_path = os.path.join(base_output_dir, file_id)
        os.makedirs(subfolder_path, exist_ok=True)

        with h5py.File(file_path, 'r') as f:
            for band in bands_to_extract:
                if band in f:
                    image = f[band][0]
                    norm_img = (255 * (image - np.min(image)) / np.ptp(image)).astype(np.uint8)

                    img = Image.fromarray(norm_img)
                    img_resized = img.resize((128, 128))

                    save_name = f"{band}_128x128.png"
                    img_resized.save(os.path.join(subfolder_path, save_name))

        print(f"‚úî Saved all bands for {file_id} in folder")
```
