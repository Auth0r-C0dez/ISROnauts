```python
import os
import glob

# Change this to the parent folder containing subfolders like 'Jun25_132271'
base_dir = '/content/drive/MyDrive/ISRO/DS/A/Jun25_132271'  # <— adjust as needed

# Find .h5 files directly under this folder
h5_paths = glob.glob(os.path.join(base_dir, '*.h5'))
print(f"Found {len(h5_paths)} .h5 files:")
for p in h5_paths:
    print(" -", p)

# Optionally store the list for later use
# For example:
all_files = h5_paths
```

```python
import h5py

# Use one file from Segment 1. For example:
file_path = all_files[0]  # or replace with a string path, e.g., '/content/drive/.../3SIMG_24JUN2025_0100_L1C_SGP_V01R00.h5'

print("Inspecting file:", file_path)
with h5py.File(file_path, 'r') as f:
    def print_structure(name, obj):
        if isinstance(obj, h5py.Dataset):
            print(f"Dataset: {name}   shape={obj.shape}   dtype={obj.dtype}")
        else:
            print(f"Group: {name}")
    f.visititems(print_structure)
```

```python
import h5py
import matplotlib.pyplot as plt
import os

# File to open (from Segment 1)
file_path = all_files[0]  # or explicitly: '/content/drive/.../3SIMG_24JUN2025_0100_L1C_SGP_V01R00.h5'

# Set this to the dataset path you found in Segment 2.
# For example: 'SomeGroup/SomeDataset'
dataset_name = None  # <— replace None with the actual string once you know it

with h5py.File(file_path, 'r') as f:
    # If user did not fill dataset_name, try auto-select:
    if dataset_name is None:
        # Collect 2D datasets:
        candidates_2d = []
        def collect_2d(name, obj):
            if isinstance(obj, h5py.Dataset) and len(obj.shape) == 2:
                candidates_2d.append(name)
        f.visititems(collect_2d)
        if candidates_2d:
            dataset_name = candidates_2d[0]
            print("Auto-selected 2D dataset:", dataset_name)
        else:
            # Try 3D
            candidates_3d = []
            def collect_3d(name, obj):
                if isinstance(obj, h5py.Dataset) and len(obj.shape) == 3:
                    candidates_3d.append(name)
            f.visititems(collect_3d)
            if candidates_3d:
                dataset_name = candidates_3d[0]
                print("Auto-selected 3D dataset:", dataset_name)
            else:
                print("No 2D or 3D datasets found. Please set dataset_name manually based on Segment 2 output.")
    if not dataset_name:
        raise ValueError("dataset_name not set and auto-selection failed.")

    print(f"Loading dataset '{dataset_name}' from file.")
    data = f[dataset_name][:]

# Decide how to handle data:
# If data is 2D, plot directly.
# If 3D, pick first slice/first band.
if data.ndim == 2:
    arr = data
elif data.ndim == 3:
    # Common patterns: (bands, H, W) or (H, W, bands). We try both:
    if data.shape[0] <= 10:
        # assume (bands, H, W)
        arr = data[0]
        print("Picked first band (index 0) from shape", data.shape)
    else:
        # assume (H, W, bands)
        arr = data[:, :, 0]
        print("Picked first channel (index 0) from shape", data.shape)
else:
    raise ValueError(f"Dataset has ndim={data.ndim}, cannot plot directly.")

# Plot with matplotlib
plt.figure(figsize=(6, 6))
plt.imshow(arr, cmap='gray')
plt.title(f"{os.path.basename(file_path)}: {dataset_name}")
plt.axis('off')
plt.colorbar(label='Value')
plt.show()
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# Load the HDF5 file
file_path = '/content/drive/MyDrive/ISRO/DS/A/Jun25_132271/3SIMG_24JUN2025_0000_L1C_SGP_V01R00.h5'  # Change this
with h5py.File(file_path, 'r') as f:

    def show_image(data, title, cmap='gray'):
        plt.figure(figsize=(8, 6))
        plt.imshow(data, cmap=cmap)
        plt.title(title)
        plt.colorbar()
        plt.axis('off')
        plt.show()

    def show_plot(data, title, kind='line'):
        plt.figure(figsize=(10, 4))
        if kind == 'bar':
            plt.bar(np.arange(len(data)), data)
        else:
            plt.plot(data)
        plt.title(title)
        plt.grid(True)
        plt.show()

    for name in f:
        dataset = f[name][()]
        print(f"Visualizing: {name} | shape={dataset.shape} | dtype={dataset.dtype}")

        if dataset.ndim == 3 and dataset.shape[0] == 1:
            # Remove singleton dimension
            show_image(dataset[0], name, cmap='inferno')

        elif dataset.ndim == 1 and dataset.shape[0] == 1024:
            show_plot(dataset, name, kind='line')

        elif dataset.ndim == 1:
            print(f"{name}: {dataset}")

        elif dataset.ndim == 0 or dataset.shape == (1,):
            print(f"{name}: {dataset.item()}")

        else:
            print(f"{name} (unusual shape): {dataset.shape}")
```
